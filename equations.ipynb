```python
# Solving normal equations using Numpy
import numpy as np 
A = np.array([[2,3],[5,4]]) 
B = np.array([8,20]) 
X = np.linalg.solve(A,B) 
X
```

```python
#Solving complex equations using numpy 
A_complex = np.array([[1+2j,2-1j],[3-1j,4]]) 
B_complex = np.array([4+1j,3+4j]) 
X_complex = np.linalg.solve(A_complex,B_complex) 
X_complex
```

```python
#vector plotting 
import matplotlib.pyplot as plt 
v = np.array([3,4]) 
plt.figure() 
plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='red') 
plt.xlim(-5,5) 
plt.ylim(-5,5) 
plt.grid() 
plt.title('vector plot') 
plt.xlabel('x-axis') 
plt.ylabel('y-axis')
plt.xticks(np.arange(-5,6,1)) 
plt.yticks(np.arange(-5,6,1)) 
plt.gca().set_aspect('equal',adjustable='box')
plt
```

```python
# Rotation of vector in 45 degree angle 
theta = np.radians(45) 
rotation = np.array([[np.cos(theta),-np.sin(theta)],
                      [np.sin(theta),np.cos(theta)]]) 
v_rotated = np.dot(rotation,v) 
plt.figure() 
plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='red',label = 'Original Vector')  
plt.quiver(0,0,v_rotated[0],v_rotated[1],angles='xy',scale_units='xy',scale=1,color='blue',label='Rotated Vector') 
plt.xlim(-5,5) 
plt.ylim(-5,5) 
plt.grid(True,which='both',linestyle='--',linewidth=0.5) 
plt.title('vector transformation plot') 
plt.xlabel('x-axis') 
plt.ylabel('y-axis') 
plt.gca().set_aspect('equal',adjustable='box')
plt.legend(loc='lower center')  
plt
```

```python
# Shear transformation
shear_factor = 1.5 
shear_matrix = np.array([[1,shear_factor],[0,1]]) 
v_sheared = np.dot(shear_matrix,v) 
plt.figure() 
plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,label='Original Vector',color='green') 
plt.quiver(0,0,v_rotated[0],v_rotated[1],angles='xy',scale_units='xy',scale=1,color='red',label='rotated Vector') 
plt.quiver(0,0,v_sheared[0],v[1],angles='xy',scale_units='xy',scale=1,label='Sheared Vector',color='blue') 
plt.xlim(-10,10) 
plt.ylim(-10,10) 
plt.grid(True,which='both',linestyle='--',linewidth=0.5) 
plt.title('Vector transformation plot') 
plt.xlabel('x-axis') 
plt.ylabel('y-axis')
plt.gca().set_aspect('equal',adjustable='box')  
plt.legend(loc='lower center') 
plt
```
